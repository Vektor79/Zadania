# Wstęp

Przed nami ostatni element modułu `Python - Analiza danych`: warsztat. Wcielimy się w rolę analityka w nowo powstałym zakładzie bukmacherskim. Naszym zadaniem jest uzyskanie wyników ligowych hokeja z publicznie dostępnych źródeł, a następnie wyznaczenie wyjściowych kursów bukmacherskich, które będą aktualizowane w miarę pojawiania się nowych danych.  
  
Pracę podzieliliśmy z kolegą z zespołu na następujące etapy:  
- **Pobranie danych** - w tym kroku pobierzemy dane ze strony [Scrape This Site](https://www.scrapethissite.com/pages/forms/) (notatnik `01_DataDownload`),  
- **Przetwarzanie danych** - dokonamy parsowania stron HTML i przygotujemy dane do analizy (notatnik `02_DataProcessing`),  
- **Analiza danych** - przeprowadzimy eksploracyjną analizę danych, aby zapoznać się z ich charakterystyką (notatnik `03_DataAnalysis`),  
- **Wydanie rekomendacji** - sformułujemy wstępne kursy bukmacherskie i omówimy możliwe sposoby na dalsze ulepszanie naszego procesu (notatnik `04_BusinessRecommendations`). 

# Wymagane biblioteki

W trakcie rozwiązywania tego warsztatu będziemy pracować przede wszystkim z następującymi bibliotekami:
1. **Selenium** - do pobrania informacji o aktualnych wynikach,
1. **BeautifulSoup** - w celu przetworzenia danych niestrukturyzowanych (kodu HTML) w dane tabelaryczne (DataSet),
1. **Pandas** - w celu wykonania transformacji danych,
1. **MatplotLib** - do prezentacji wyników.

## Co powtórzyć przed warsztatem
Warsztat będzie bazował na wiedzy z całego kursu. Jednak niektóre rzeczy warto sobie powtórzyć. Są to:

- Pobieranie kodu strony za pomocą Selenium i zapis tego kodu do pliku.
- Odczytywanie zapisanego na dysku pliku.
- Przeszukiwanie folderów (moduł glob omawiany podczas zajęć z Pythona).
- Iterowanie po strukturach danych (listy, słowniki itp.).
- Wyszukiwanie elementów w kodzie strony za pomocą BeautifulSoup.

# Struktura projektowa

Aby ułatwić sobie pracę w trakcie rozwiązywania warsztatu, przyjmiemy pewne założenia co do struktury projektu. Praca z tą samą strukturą pozwoli na uniknięcie dodatkowej pracy przy współdzieleniu rozwiązania.

## data
W tym folderze znajdować się będą wszystkie dane, które wytworzysz w trakcie rozwiązywania zadania.

### data/raw (dane źródłowe)

Tutaj umieścimy surowe pliki `.html`, pobrane ze strony. Będą to pliki nieprzetworzone, bez żadnej dodatkowej obróbki - innymi słowy czysta kopia html.

> Pomijając kontekst projektu, często pliki źródłowe przychodzą w formie plików Excel, które łatwo ręcznie zmodyfikować. Dobra praktyka mówi o tym, że pliki źródłowe są nieedytowalne tj., wszystkie zmiany powinny być w skryptach, w celu późniejszej replikacji pracy.

### data/interim (dane tymczasowe)
Tutaj będziemy zamieszczać wynik parsowania danych z folderu **raw** do postaci listy `json`. Dopiero dane z tego folderu zostaną przekazane dalej w celu obróbki/transformacji.

### data/processed (dane wynikowe)
Ten folder będzie zawierał dane wynikowe - oczyszczone, gotowe do dalszej analizy. W kontekście warsztatu to na ich podstawie będziemy wyciągać wnioski oraz rekomendacje.

## drivers
Wymagane sterowniki do użycia Selenium

## notebooks
Tutaj będziemy zamieszczać wszystko notebooki stworzone w trakcie warsztatu. Ich nazewnictwo powinno być następując <krok>-<opis>.ipynb, dzięki temu w momencie próby odtworzenia warsztatu będzie znana kolejność wykonywanych kroków.

## venv (opcjonalny)
Folder zawierający wirtualne środowisko python'a wraz z zainstalowanymi tylko tymi paczkami, które są potrzebne do jego poprawnego działania. Opis tworzenia jest dodatkowo opisany w materiałach po zjeździe.

## Dodatkowy komentarz
Zaprezentowana struktura nazywa się cookiecutter dla Data Science, jednak aby nie wprowadzać dodatkowe zamieszania, została uproszczona na potrzeby warsztatowe. Więcej informacji na temat tego podejścia można znaleźć [tutaj](https://cookiecutter-data-science.drivendata.org/).
