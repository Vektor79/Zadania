# Pierwszy wykres

Uruchom poniższy kod, żeby sprowadzić, czy masz poprawnie zainstalowane i skonfigurowane wszystkie potrzebne biblioteki.

````python
import matplotlib.pyplot as plt


x = [0,1,2,3,4,5]
y1 = [0,1,2,3,4,5]
y2 = [0,1,4,9,16,25]


plt.plot(x, y1, label='Wykres 1')
plt.plot(x, y2, label='Wykres 2')  
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()

plt.show();
````

Jeżeli wykres Ci się nie pojawia w powyżej komórce dopisz:
